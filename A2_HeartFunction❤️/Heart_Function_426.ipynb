```python
import numpy as np
import matplotlib.pyplot as plt

class HeartCurveGenerator:
    """
    A class to generate the coordinates for a heart-shaped curve using two methods:
    1. Parametric Equations
    2. Explicit Formula
    """
    def __init__(self):
        """
        Initializes the HeartCurveGenerator class.
        """
        self.t = np.linspace(0, 2 * np.pi, 1000)  # For parametric equations
        self.x_data = np.linspace(-1, 1, 1000)    # For explicit formula

    def generate_parametric_curve(self):
        """
        Generates the x and y coordinates for the heart curve using parametric equations.
        """
        x = 16 * np.sin(self.t)**3
        y = 13 * np.cos(self.t) - 5 * np.cos(2*self.t) - 2 * np.cos(3*self.t) - np.cos(4*self.t)
        return x, y

    def generate_explicit_curve(self):
        """
        Generates the x and y coordinates for the heart curve using the explicit formula.
        """
        y_upper = np.sqrt(1 - self.x_data**2) + np.abs(self.x_data)**(2/3)
        y_lower = -np.sqrt(1 - self.x_data**2) + np.abs(self.x_data)**(2/3)
        return self.x_data, y_upper, y_lower

class HeartCurvePlotter:
    """
    A class to plot the heart-shaped curves.
    """
    def __init__(self):
        """
        Initializes the HeartCurvePlotter class.
        """
        pass

    def plot_parametric_curve(self, x, y):
        """
        Plots the parametric heart curve with customization.
        """
        plt.figure(figsize=(6, 6))
        plt.plot(x, y, 'r', linewidth=2)
        plt.axhline(0, color='black', linewidth=1)  # X-axis
        plt.axvline(0, color='black', linewidth=1)  # Y-axis
        plt.grid(True, linestyle='--', alpha=0.6)  # Grid
        plt.xlim(-20, 20)
        plt.ylim(-20, 20)
        plt.xticks(np.arange(-20, 21, 5))
        plt.yticks(np.arange(-20, 21, 5))
        plt.title("Parametric Heart Shape ❤️", fontsize=14)
        plt.legend()
        plt.show()

    def plot_explicit_curve(self, x, y_upper, y_lower):
        """
        Plots the explicit formula heart curve with customization.
        """
        plt.figure(figsize=(6, 6))
        plt.plot(x, y_upper, 'b', linewidth=2)
        plt.plot(x, y_lower, 'b', linewidth=2)
        plt.axhline(0, color='black', linewidth=1)  # X-axis
        plt.axvline(0, color='black', linewidth=1)  # Y-axis
        plt.grid(True, linestyle='--', alpha=0.6)  # Grid
        plt.xlim(-1.5, 1.5)  # Adjusted for explicit formula range
        plt.ylim(-2, 2)      # Adjusted for explicit formula range
        plt.xticks(np.arange(-1.5, 1.6, 0.5))
        plt.yticks(np.arange(-2, 2.1, 0.5))
        plt.title("Explicit Formula Heart Shape ❤️", fontsize=14)
        plt.legend()
        plt.show()

# Main execution
if __name__ == "__main__":
    # Step 1: Generate the heart curve coordinates
    heart_curve_generator = HeartCurveGenerator()
    x1, y1 = heart_curve_generator.generate_parametric_curve()
    x2, y2_upper, y2_lower = heart_curve_generator.generate_explicit_curve()

    # Step 2: Plot the heart curves
    heart_plotter = HeartCurvePlotter()

    # Plot parametric curve
    heart_plotter.plot_parametric_curve(x1, y1)

    # Plot explicit formula curve
    heart_plotter.plot_explicit_curve(x2, y2_upper, y2_lower)
```
