```python
import numpy as np
import matplotlib.pyplot as plt
"""
    Generates and plots a heart-shaped curve using parametric equations. The function uses trigonometric functions to create the x and y coordinates
    of the heart shape.

"""
class HeartCurveGenerator:
    """
    A class to generate the coordinates for a heart-shaped curve.
    """
    def __init__(self):
        """
        Initializes the HeartCurveGenerator class.
        """
        self.t = np.linspace(0, 2 * np.pi, 1000)  # Generate 1000 points between 0 and 2*pi

    def generate_curve(self):
        """
        Generates the x and y coordinates for the heart curve.
        """
        x = 16 * np.sin(self.t)**3
        y = 13 * np.cos(self.t) - 5 * np.cos(2*self.t) - 2 * np.cos(3*self.t) - np.cos(4*self.t)
        return x, y

class HeartCurvePlotter:
    """
    A class to plot the heart-shaped curve.
    """
    def __init__(self, x, y):
        """
        Initializes the HeartCurvePlotter class with x and y coordinates.
        """
        self.x = x
        self.y = y

    def plot(self):
        """
        Plots the heart curve with customization.
        """
        plt.figure(figsize=(6, 6))
        plt.plot(self.x, self.y, 'r', linewidth=2)  # Plot the curve
        plt.axhline(0, color='black', linewidth=1)  # X-axis
        plt.axvline(0, color='black', linewidth=1)  # Y-axis
        plt.grid(True, linestyle='--', alpha=0.6)  # Grid
        plt.xlim(-20, 20)
        plt.ylim(-20, 20)
        plt.xticks(np.arange(-20, 21, 5))
        plt.yticks(np.arange(-20, 21, 5))
        plt.title("Heart Shape ❤️", fontsize=14)
        plt.show()

# Main execution
if __name__ == "__main__":
    # Step 1: Generate the heart curve coordinates
    heart_curve_generator = HeartCurveGenerator()
    x, y = heart_curve_generator.generate_curve()

    # Step 2: Plot the heart curve
    heart_plotter = HeartCurvePlotter(x, y)
    heart_plotter.plot()
```
